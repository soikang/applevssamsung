compare with the words that are removed special chars and stopwords. Any difference?

```python
print(data.iloc[13210:,])

reviews = data['review']
stopwords = nltk.corpus.stopwords.words('english')

print(type(reviews))
print(reviews.iloc[13210:,])

# df with one column
df = reviews.to_frame()
print(len(df['review']))

# remove non ASCI characters
df['cleaned'] = df['review'].str.replace(r'[^\x00-\x7f]', '')
print(df.iloc[13210:,])

# remove special chars
df['cleaned'].replace(to_replace=r'[?|$|.|!|;|:|(|)]', value='',inplace=True, regex=True) 

print(len(df['cleaned']))
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

word_tokens = []

for review in df['cleaned'] :
    tokens = word_tokenize(str(review))
    word_tokens.append(tokens)

filtered_sentence = []
for tokens in word_tokens :
    
    filtered_review = [w for w in tokens if not w in stopwords] 
    filtered_sentence.append(filtered_review)
```

```python
print(df['cleaned'].iloc[13210:])


filtered = pd.DataFrame({'filtered' : filtered_sentence})
print(len(filtered))

print(filtered.iloc[13210:])

df['filtered'] = filtered
print(len(filtered))
```

```python
df['filtered'] = df['filtered'].apply(' '.join)
```

```python
# polarity, subjectivity

df
```

```python
# get the scores
df['polarity_review'] = df['review'].apply(lambda x : TextBlob(x).sentiment[0])
df['polarity_filtered'] = df['filtered'].apply(lambda x : TextBlob(x).sentiment[0])
df['subjectivity_review'] = df['review'].apply(lambda x : TextBlob(x).sentiment[1])
df['subjectivity_filtered'] = df['filtered'].apply(lambda x : TextBlob(x).sentiment[1])
```

```python
len(df)
```

```python
import numpy as np

print("original review's polarity sum :" + str(np.sum(df['polarity_review'])))
print("filtered review's polarity sum :" + str(np.sum(df['polarity_filtered'])))
print()
print("original review's subjectivity sum : " + str(np.sum(df['subjectivity_review'])))
print("filtered review's subjectivity sum : " + str(np.sum(df['subjectivity_filtered'])))
```
